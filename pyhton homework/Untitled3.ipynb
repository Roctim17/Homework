```python
import pandas as pd

# 示例 DataFrame
data = {
    'age': [23, 45, 28, 52, 34, 38, 41, 29],
    'salary': [30000, 80000, 40000, 90000, 50000, 60000, 70000, 45000]
}
example3_1 = pd.DataFrame(data)

# 计算所有数值列的分位数
result = example3_1.quantile(q=[0.25, 0.75])
print(result)
```

```python
import pandas as pd

# 创建示例数据
example3_1 = pd.DataFrame({
    'love': [80, 90, 75, 95, 85],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve']
})

# 计算 'love' 列的极差（最大值 - 最小值）
R = example3_1['love'].max() - example3_1['love'].min()

print('humen:', R)
```

```python
import numpy as np
import pandas as pd

# 创建示例数据
example3_1 = pd.DataFrame({
    'love': [80, 90, 75, 95, 85, 70, 100]
})

# 计算 IQR（四分位距）
IQR = np.quantile(example3_1['love'], q=0.75) - np.quantile(example3_1['love'], q=0.25)

print('IQR =', IQR)
```

```python
import pandas as pd

# 创建示例数据
example3_1 = pd.DataFrame({
    'love': [80, 90, 75, 95, 85, 70, 100]
})

# 计算样本方差（ddof=1 表示自由度为 1，即除以 n-1）
var = example3_1['love'].var(ddof=1)

print('human:', var)
```

```python
from scipy import stats
import pandas as pd

# 示例数据
example3_1 = pd.DataFrame({
    'cereness': [80, 90, 75, 95, 85]
})

# 计算 Z-score，并设置 ddof=1（可选，zscore 中 ddof 不常用，但支持）
z = stats.zscore(example3_1['cereness'], ddof=1)

print(z)
```

```python
import pandas as pd

# Step 1: 创建 example3（模拟数据）
example3 = pd.DataFrame({
    '分数': [85, 92, 78, 96, 88, 79, 94, 87, 91, 83]
})

# Step 2: 计算偏度和峰度
skew = example3['分数'].skew()
kurt = example3['分数'].kurt()

# Step 3: 打印结果
print("偏度系数:", round(skew, 4))
print("峰度系数:", round(kurt, 4))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 模拟数据（用于测试）
example3_12 = pd.DataFrame({
    '月生活费支出': [1000, 1500, 1200, 1800, 2000, 900, 1300, 1600, 1400, 1700],
    '性别': ['男', '女', '女', '男', '女', '男', '女', '男', '女', '男'],
    '家庭所在地': ['城市', '农村', '城市', '农村', '城市', '城市', '农村', '城市', '农村', '城市']
})

# 绘图
plt.figure(figsize=(12, 5))

# 子图1：按性别
plt.subplot(1, 2, 1)
sns.kdeplot(data=example3_12, x='月生活费支出', hue='性别', shade=True)
plt.title('(a) 按性别分组')

# 子图2：按家庭所在地
plt.subplot(1, 2, 2)
sns.kdeplot(data=example3_12, x='月生活费支出', hue='家庭所在地', shade=True)
plt.title('(b) 按家庭所在地分组')

# 调整布局
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 绘制分组箱线图：按性别和家庭所在地比较月生活费支出
sns.boxplot(
    data=example3_12,
    x='性别',
    y='月生活费支出',
    hue='家庭所在地'
)

# 显示图形
plt.title('按性别和家庭所在地分组的月生活费支出分布')
plt.xlabel('性别')
plt.ylabel('月生活费支出')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 5))

# 子图1：按性别
plt.subplot(1, 2, 1)
sns.boxplot(data=example3_12, x='性别', y='月生活费支出')
plt.title('(a) 按性别分组')

# 子图2：按家庭所在地（散点图）
plt.subplot(1, 2, 2)
sns.stripplot(data=example3_12, x='家庭所在地', y='月生活费支出', size=5)
plt.title('(b) 按家庭所在地分组')

# 调整布局并显示
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 假设 example3_12 已定义
result = pd.crosstab(
    index=example3_12['性别'],
    columns=example3_12['家庭所在地'],
    margins=True,
    margins_name='合计'
)

print(result)
```

```python
import pandas as pd
import numpy as np

tab = pd.pivot_table(
    data=example3_12,
    index=['性别', '家庭所在地'],
    values=['月生活费支出'],
    aggfunc='mean'  # 可改为 'sum', 'count', 或多个函数
)

print(tab)
```
