```python
import pandas as pd

s = pd.Series(['a', 'b', 'c', 'd'])
# Insert 'dfjkd' at index 2
s_new = pd.concat([
    s.iloc[:2],                    # First 2 elements
    pd.Series(['dfjkd']),          # Value to insert
    s.iloc[2:]                     # Remaining elements
], ignore_index=False)

print(s_new)
```
