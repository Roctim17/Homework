```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据
y = np.array([1, 2, 3, 4, 5])

plt.figure()
plt.plot(y, marker='o')  # 绘制数据点

# 添加水平参考线（均值）
plt.axhline(y.mean(), color='black', ls='--', lw=1)

plt.xlabel('X轴')
plt.ylabel('Y轴')
plt.title('带均值线的折线图')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 8))

# 子图 1
plt.subplot(2, 2, 1)
plt.hist(np.random.normal(loc=0, scale=1, size=100), bins=20)
plt.title('μ=0, σ=1')

# 子图 2 ✅ 你正在写的
plt.subplot(2, 2, 2)
plt.hist(np.random.normal(loc=5, scale=10, size=100), bins=20)
plt.title('μ=5, σ=10')

# 子图 3
plt.subplot(2, 2, 3)
plt.hist(np.random.normal(loc=10, scale=5, size=100), bins=20)
plt.title('μ=10, σ=5')

# 子图 4
plt.subplot(2, 2, 4)
plt.hist(np.random.normal(loc=-5, scale=15, size=100), bins=20)
plt.title('μ=-5, σ=15')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建第2个子图（2x2布局中的第2个）
plt.subplot(2, 2, 2)

# 生成正态分布数据并绘制直方图
plt.hist(
    np.random.normal(loc=5, scale=10, size=100),
    bins=20,
    alpha=0.7,
    color='blue',
    edgecolor='black'
)

plt.title('Normal Distribution: μ=5, σ=10')
```

```python
import matplotlib.pyplot as plt

# 创建子图 ax1
ax1 = plt.subplot2grid((3, 3), (0, 0), rowspan=1, colspan=1)
ax1.plot([1, 2, 3], [1, 4, 2])
ax1.set_title('ax1')

# 创建子图 ax2：从 (0,2) 开始，跨3行，1列
ax2 = plt.subplot2grid((3, 3), (0, 2), rowspan=3, colspan=1)
ax2.bar([0, 1, 2], [3, 7, 5])
ax2.set_title('ax2')

# 调整布局
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 6))

# 创建 GridSpec
spec = fig.add_gridspec(
    nrows=2,
    ncols=6,
    width_ratios=[1, 1, 1, 1, 1, 1],  # 必须有 6 个值，对应 6 列
    height_ratios=[1, 2]              # 2 个值，对应 2 行
)

# 示例：添加子图
ax1 = fig.add_subplot(spec[0, 0:2])  # 第0行，第0-1列
ax1.plot([1, 2, 3], [1, 4, 2])
ax1.set_title('ax1')

ax2 = fig.add_subplot(spec[:, 2:])   # 所有行，第2-5列
ax2.bar([0, 1, 2], [3, 7, 5])
ax2.set_title('ax2')

plt.show()
```
