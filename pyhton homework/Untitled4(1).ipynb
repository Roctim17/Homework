```python
import pandas as pd

# 创建一个示例数据
data = {
    'score': [88, 92, 76, 95, 85, 80, 90],
    'age': [20, 21, 19, 22, 20, 19, 21]
}

# 定义 example3_1 这个变量（一个 DataFrame）
example3_1 = pd.DataFrame(data)

# 然后调用 quantile 方法
result = example3_1.quantile(q=[0.25, 0.75], interpolation='linear')
print(result)
```

```python
import pandas as pd  # ✅ 拼写正确

# 假设你已经定义了 example3_1（例如从数据中读取）
# 如果还没有定义，请先创建它，比如：
# example3_1 = pd.read_csv('your_data.csv')  # 或手动创建
data = {'heart': [1, 2, 2, 3, 2], 'blood_pressure': [80, 90, 90, 85, 90]}
example3_1 = pd.DataFrame(data)

# 计算某一列的众数
mode_heart = example3_1['heart'].mode()

# 计算整个 DataFrame 每一列的众数
mode_all = example3_1.mode()

print("heart列的众数：", mode_heart.values)  # 输出数值
print("所有列的众数：\n", mode_all)
```

```python
import pandas as pd  # ✅ 正确拼写

# 必须先定义 example3_1，例如：
data = {
    'medicine': [10, 25, 15, 30, 20],
    'patient': ['A', 'B', 'C', 'D', 'E']
}
example3_1 = pd.DataFrame(data)  # ✅ 创建 DataFrame

# 计算 medicine 列的极差（最大值 - 最小值）
R = example3_1['medicine'].max() - example3_1['medicine'].min()

# 输出结果
print('range:', R)  # 建议使用 'range' 而不是 'neuclear'
```
